### MINIO 概念
自己部署的“私有版S3对象存储“   
#### 对象存储概念
##### 什么是S3?
来自Amazon Web Services， 全称 Simple Storage Service （现在已经是一个行业标准接口） 
##### S3结构
三层
1，bucket 桶；
2，Object 对象；
包含文件内容(data)，元数据(metadata)，唯一的key(Path)
3, Key 
#### S3特点
1 可扩展（无限容量）。  
2 高可靠 （数据自动备份）  
3 HTTP访问。 
4 权限控制。
